## 1 Prelab

i = 1/2 * Kn * Vov^2

Vgs = Vds

i = (Vdd - Vds) / (Rref)

Vds = Vdd - i*Rref = Vgs

i = 1/2 * Kn * (Vdd - i * R - Vgs) ^ 2

i = 1/2 * Kn * [(Vdd^2 - Vdd * R - Vth*Vdd) + (-Vdd * i * R + i^2 * R^2 + Vth * i * R) + (-Vth * Vdd + Vth * i * R + Vthn^2)]

i = 1/2 * Kn[Vdd^2 + i^2 * R^2 + Vth ^ 2 - 2 * Vdd * i * R + 2 * Vth * i * R - 2 * Vth * Vdd]

0 = i^2(1/2 * Kn * R^2) + i(-2 * Vdd * R + 2 * Vth * R - 1) + (Vdd^2 + Vth^2 - 2 * Vth * Vdd)

i = ((2 * Vdd * r - 2 * Vth * R + 1) +- sqrt[(-2 * Vdd * R + 2 * Vth * r - 1)^2 - 2 * K * R^2 * (Vdd ^ 2 + Vth ^ 2 - 2 * Vth * Vdd)]) / (Kn * R)


````python
from math import sqrt
Vdd = 5 # V
Vth = 0.75
Kn = 0.7e-3 # A
R = 10e3 # Ohm

i_refp = ((2 * Vdd * R - 2 * Vth * R + 1) + sqrt((-2 * Vdd * R + 2 * Vth * R - 1)**2 - 2 * Kn * R**2 * (Vdd**2 + Vth**2 - 2 * Vth * Vdd))) / (Kn * R)
i_refn = ((2 * Vdd * R - 2 * Vth * R + 1) - sqrt((-2 * Vdd * R + 2 * Vth * R - 1)**2 - 2 * Kn * R**2 * (Vdd**2 + Vth**2 - 2 * Vth * Vdd))) / (Kn * R)

print(i_refp)
print(i_refn)
````
